```python
import concurrent.futures
import requests
import time

# Function to check a single link
def check_link(link):
    try:
        response = requests.get(link)
        if response.status_code != 200:
            print(f"Broken link found: {link}")
        else:
            print(f"Link is okay: {link}")
    except requests.exceptions.RequestException as e:
        print(f"Error occurred for link: {link}, Error: {e}")

# List of links to check
links = [
    "https://www.example.com",
    "https://www.google.com",
    "https://www.openai.com",
    "https://www.nonexistentlink12345.com",
]

# Dictionary to store execution times
execution_times = {}
rep=40
ammountoflinks=len(links)*rep

# Loop over different thread counts
for max_threads in range(1,5):
    start_time = time.time()

    with concurrent.futures.ThreadPoolExecutor(max_threads) as executor:
        futures = [executor.submit(check_link, link) for link in links*rep]

        # Wait for all tasks to complete
        concurrent.futures.wait(futures)

    execution_time = time.time() - start_time
    execution_times[max_threads] = execution_time

    print(f"--- {max_threads} threads: {execution_time} seconds ---")

print("Link checking completed.")
print(f"Time to check {ammountoflinks} links")
print("Execution times:", execution_times)
```
